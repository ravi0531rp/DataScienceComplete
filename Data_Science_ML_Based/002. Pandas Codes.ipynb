### Apply on Multiple Columns

```python
df['total_bill'].apply(lambda x: x**2)
```

```python
qualityOfTip = lambda tot,tip : 'Generous' if tip/tot >= 0.25 else 'Not Generous'
```

```python
qualityOfTip(16.99,1.2)
```

```python
lambda df : qualityOfTip(df['total_bill'] , df['tip'])
```

```python
x =df[['total_bill','tip']].apply(lambda df : qualityOfTip(df['total_bill'] , df['tip']) , axis = 1)
```

```python
type(x)
```

```python
x
```

### Here it did 2 things:
* The lambda fn inside apply simply fetched the columns from df and applied the other aggregate fn
* The other aggregate fn did the computation

### Other way : Vectorize

```python
x = np.vectorize(qualityOfTip)(df['total_bill'],df['tip'])  # It is seriously fast here, not always!!!!
```

```python
type(x)
```
