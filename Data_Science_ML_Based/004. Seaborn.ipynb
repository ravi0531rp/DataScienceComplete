# Scatter Plots

Scatter plots can show how different features are related to one another, the main theme between all relational plot types is they display how features are interconnected to each other. There are many different types of plots that can be used to show this, so let's explore the scatterplot() as well as general seaborn parameters applicable to other plot types.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('D:\\ALL_CODES_SECTIONWISE\\DSC_CODES\\Data_Science_ML_Based'+"\\DATA\\dm_office_sales.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(12,4) , dpi=200)
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(12,4) , dpi=200)
sns.scatterplot(x='salary',y='sales',data=df , hue='level of education') # hue : color by this ; categorical -> separate colors
```

```python
plt.figure(figsize=(12,4) , dpi=200)
sns.scatterplot(x='salary',y='sales',data=df , hue='salary') # hue : color by this column ; continuous -> color gradient
```

```python
sns.scatterplot(x='salary',y='sales',data=df , hue='salary' , palette='Dark2') 
```

```python
sns.scatterplot(x='salary',y='sales',data=df , hue='salary' , size = 'salary') 
```

```python
sns.scatterplot(x='salary',y='sales',data=df , hue='level of education' , style = 'level of education') 
plt.savefig("imgs//scatter_complex.jpg",bbox_inches='tight')
```
