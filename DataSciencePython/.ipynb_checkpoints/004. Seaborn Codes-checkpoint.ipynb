# Scatter Plots

Scatter plots can show how different features are related to one another, the main theme between all relational plot types is they display how features are interconnected to each other. There are many different types of plots that can be used to show this, so let's explore the scatterplot() as well as general seaborn parameters applicable to other plot types.

```python
import pandas as pd
import seaborn as sns
```

```python
pwd
```

```python
df = pd.read_csv('D:\\ALL_CODES_SECTIONWISE\\DSC_CODES\\Data_Science_ML_Based'+"\\DATA\\dm_office_sales.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(12,4) , dpi=200)
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(12,4) , dpi=200)
sns.scatterplot(x='salary',y='sales',data=df , hue='level of education') # hue : color by this ; categorical -> separate colors
```

```python
plt.figure(figsize=(12,4) , dpi=200)
sns.scatterplot(x='salary',y='sales',data=df , hue='salary') # hue : color by this column ; continuous -> color gradient
```

```python
sns.scatterplot(x='salary',y='sales',data=df , hue='salary' , palette='Dark2') 
```

```python
sns.scatterplot(x='salary',y='sales',data=df , hue='salary' , size = 'salary') 
```

```python
sns.scatterplot(x='salary',y='sales',data=df , hue='level of education' , style = 'level of education') 
plt.savefig("imgs//scatter_complex.jpg",bbox_inches='tight')
```

# Understand Plot Types

```python
sns.rugplot(x='salary' , data=df , height = 0.5) # yaxis values don't mean anything here
```

```python
## Do not use sns.distplot ; use dis or his
```

```python
sns.set(style='darkgrid') # sets the style for the notebook itself
sns.displot(data = df, x='salary', bins=20 , color='red' , edgecolor='blue' , kde=True , rug = True)
```

```python
sns.histplot(data = df, x='salary', bins=20 , color='red' , edgecolor='blue')
```

```python
sns.kdeplot(data=df, x = 'salary')
```

```python
df = pd.read_csv('D:\\ALL_CODES_SECTIONWISE\\DSC_CODES\\Data_Science_ML_Based\\DATA\\country_table.csv')
```

```python
df = df.set_index('Countries')
```

```python
df
```

```python
sns.heatmap(df) # problem here is that all except Life Expectancy is a rate value
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1) , linewidth=0.5, annot = True , cmap = 'viridis')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1).corr() , linewidth=0.5, annot = True , cmap = 'viridis')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1) , linewidth=0.5, annot = True , cmap = 'viridis')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1) , linewidth=0.5, annot = True , cmap = 'viridis' , col_cluster=False)
```
