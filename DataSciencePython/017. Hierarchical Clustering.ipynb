```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DATA/cluster_mpg.csv')
```

```python
df_new = pd.get_dummies(df.drop('name',axis=1))
```

```python
df_new.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_new)
```

```python
scaled_data.head()
```

```python
scaled_df = pd.DataFrame(scaled_data , columns=df_new.columns)
```

```python
scaled_df.head()
```

```python
plt.figure(figsize=(15,8) , dpi=200)
sns.heatmap(scaled_df , cmap='viridis')
```

```python
plt.figure(figsize=(15,8) , dpi=200)
sns.clustermap(scaled_df) # by default, seaborn clusters via row as well as columns!!!
```

```python
help(sns.clustermap)
```

```python
plt.figure(figsize=(15,8) , dpi=200)
sns.clustermap(scaled_df, row_cluster=False) # by default, seaborn clusters via row as well as columns!!!
```

```python
plt.figure(figsize=(15,8) , dpi=200)
sns.clustermap(scaled_df, col_cluster=False) # by default, seaborn clusters via row as well as columns!!!
```

## Coding through Scikit

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
help(AgglomerativeClustering)
```

```python
model = AgglomerativeClustering(n_clusters=5)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df, x='mpg' , y='weight' , hue= cluster_labels , palette='viridis')
```

```python
len(scaled_df.columns)
```

```python
np.sqrt(10) # max distance possible between 2 points (after using min max scaling)
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold = 0) # lol , would make no clusters!!
```

```python
cL = model.fit_predict(scaled_df)
```

```python
cL
```

```python
len(scaled_df)
```

```python
len(np.unique(cL)) # see , every point is its own cluster as thresh is 0
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(12,4), dpi=200)
dendro = dendrogram(linkage_matrix)
```
