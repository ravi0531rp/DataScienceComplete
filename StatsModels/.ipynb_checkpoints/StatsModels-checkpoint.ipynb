```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = sm.datasets.get_rdataset("Guerry" , "HistData").data
```

```python
data.head()
```

# Fit OLS regression with R style formulas

```python
data.columns
```

```python
res = smf.ols("Lottery ~ Literacy + np.log(Pop1831)" , data = data).fit()
```

```python
res.summary()
```

# Fit OLS with numpy arrays

```python
X = data["Literacy"]
y = data["Infanticide"]
```

```python
res2 = sm.OLS(y, X).fit()
```

```python
res2.summary()
```
